```python
import matplotlib.pyplot as plt
%matplotlib inline
#by default
```

```python
import numpy as np
import pandas as pd
```

```python
subject_name=['DA','ML','DBMS','SQL']
class_Month1=[10,50,7,20]
class_Month2=[5,8,21,64]
class_Month3=[61,18,9,6]
plt.plot(subject_name,class_Month1)
#plt.show()#no need to write
```

```python
plt.plot(subject_name,class_Month1,color='r',linestyle='-.',linewidth=4)
plt.xlabel('subject_name')
plt.ylabel('class_Month1')
plt.title("class list")
```

```python
#bar chart
plt.bar(subject_name,class_Month1,color='r')
plt.xlabel('subject_name')
plt.ylabel('class_Month1')
plt.title("class list")
```

```python
plt.bar(10,20,30)
```

```python
plt.bar(subject_name,class_Month1,color='m',align='edge')
plt.xlabel('subject_name')
plt.ylabel('class_Month1')
plt.title("class list")
```

```python
plt.bar(subject_name,class_Month1,class_Month2,class_Month3,color=['r','g','m'])
plt.xlabel('subject_name')
plt.ylabel('class_Month1')
plt.title("class list")
```

```python
plt.bar(subject_name,class_Month1,color='g')
plt.bar(subject_name,class_Month2,color='r')
plt.bar(subject_name,class_Month3,color='black')
plt.xlabel('subject_name')
plt.ylabel('class_Month1')
plt.title("class list")
```

```python
plt.barh(subject_name,class_Month1,color='g')
plt.barh(subject_name,class_Month2,color='r')
plt.barh(subject_name,class_Month3,color='black')
plt.xlabel('subject_name')
plt.ylabel('class_Month1')
plt.title("class list")
```
